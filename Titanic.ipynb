Importing neccesary libraries 

```python
import pandas as pd 
import numpy as np
import matplotlib as plt 
import seaborn as sns 
```

show the data 

```python
data= pd.read_csv('titanic_data.csv')
data.head()
```

```python
data.info()
```

as shown above there are mull values in Age,Cabin and Embarked 

```python
age_mean=data['Age'].mean()
print(age_mean)
```

```python
NewAgeMean=round(age_mean)
print(NewAgeMean)
```

replace th null values in Age Coloumn with the mean 

```python
data['Age'].fillna(value=NewAgeMean, inplace=True)
```

```python
data.info()
```

we don't need the cabin coloumn so we drop it 

```python
data.drop(["Cabin"], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

as we shown above there are two rows contain null values so we drop it  

```python
data = data.dropna()
```

```python
data.info()
```

Now the data is clean and does not contain any null values 

Sort the data 

```python
data.sort_values(by=['Age'])
```

```python
gender={'male':0,'female':1}
data['Sex']=data['Sex'].map(gender)
data.head()
```

Data Visualization 

```python
sns.countplot(x='Survived',data=data,hue='Sex')
```

```python
sns.countplot(x='Survived',data=data)
```

```python
data['Age'].hist(bins=10)
plt.grid(False)
```

```python
import matplotlib.pyplot as plt
sizes= data['Survived'].value_counts()
fig1,ax1 = plt.subplots()
ax1.pie(sizes,labels=['Not Survived','Survived'],autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
data.groupby(by='Survived')['Age'].describe()
```
