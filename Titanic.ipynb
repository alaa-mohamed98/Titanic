```python
import pandas as pd 
import numpy as np
import matplotlib as plt 
import seaborn as sns 
```

```python
data= pd.read_csv('titanic_data.csv')
data.head()
```

```python
data.info()
```

```python
age_mean=data['Age'].mean()
print(age_mean)
```

```python
NewAgeMean=round(age_mean)
print(NewAgeMean)
```

```python
data['Age'].fillna(value=NewAgeMean, inplace=True)
```

```python
data.info()
```

```python
data.drop(["Cabin"], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
data.sort_values(by=['Age'], ascending=[True])
```

```python
data.loc[data.loc[:, 'Age']==0.42, 'Age'] = 42
```

```python
data.loc[data.loc[:, 'Age']==0.67, 'Age'] = 67
data.loc[data.loc[:, 'Age']==0.75, 'Age'] = 75
data.loc[data.loc[:, 'Age']==0.83, 'Age'] = 83
data.loc[data.loc[:, 'Age']==0.92, 'Age'] = 29
```

```python
data.sort_values(by=['Age'])
```

```python
gender={'male':0,'female':1}
data['Sex']=data['Sex'].map(gender)
data.head()
```

Data Visualization 

```python
sns.countplot(x='Survived',data=data,hue='Sex')
```

```python
sns.countplot(x='Survived',data=data)
```

```python
data['Age'].hist(bins=10)
plt.grid(False)
```

```python
import matplotlib.pyplot as plt
sizes= data['Survived'].value_counts()
fig1,ax1 = plt.subplots()
ax1.pie(sizes,labels=['Not Survived','Survived'],autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
data.groupby(by='Survived')['Age'].describe()
```

```python
g = sns.FacetGrid(data, col='Survived')
g.map(plt.hist, 'Age', bins=20)
```
